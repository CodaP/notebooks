```python
import random
from itertools import takewhile, chain
from itertools import product as iter_product
from collections import Counter
%pylab inline
figsize(15,5)
import pandas as pd
import numpy as np
```

Part 1
======

```python
def run_walk(size=100,start=50):
    """
    Generate cell numbers for a single simulation
    """
    x = start
    while 0 <= x < size:
        yield x
        x += random.choice([-1,1])
```

1)
--

```python
samples = 10000
pdf = pd.Series([sum(1 for x in run_walk()) for i in xrange(samples)])
```

```python
pdf.hist(normed=True, bins=100)
title('Probability Density of Termination')
ylabel('Probability Density')
_ = xlabel('t (number of steps before termination)')
minorticks_on()
```

2)
--

```python
# Run simulation
samples = 10000
cell_visits = Counter([])
for i in range(samples):
    cell_visits.update(x for x in run_walk(100,50))
cell_visits = pd.Series(cell_visits)
```

```python
# Plot results
cell_visits.plot()
title('Cell visits in {0:d} samples'.format(samples))
ylabel('Visits')
_ = xlabel('Cell number')
```

##Part 2: 2-D Random Walks

```python
def run_walk_2d(size=5,start=(3,3)):
    """
    Generate cell numbers for a single simulation
    """
    x = list(start)
    while 0 <= x[0] < size and 0 <= x[1] < size:
        yield tuple(x)
        x[random.choice([0,1])] += random.choice([-1,1])
```

```python
samples = pd.DataFrame([coord for coord in chain(*[run_walk_2d(size=25,start=(12,12)) for x in range(1000)])],columns=['x','y'])
```

```python
t_length = pd.DataFrame([sum(1 for x in run_walk_2d(size=25,start=(12,12))) for i in xrange(10000)])
```

```python
t_length.hist(normed=True,bins=100)
title('Probability Density of Termination')
ylabel('Probability Density')
xlabel('t (number of steps before termination)')
vlines(t_length.median(),*ylim(),label='Median = {}'.format(t_length.median()[0]), linewidth=3)
_ = legend()
```

```python
def run_walk_2d_bounce(size=5,start=(3,3)):
    """
    Generate cell numbers for a single simulation
    """
    x = np.array(start)
    all_cells = set(iter_product(range(size),repeat=2))
    visited = set([])
    while visited != all_cells:
        x = np.clip(x,0,size-1)
        visited.add(tuple(x))
        yield tuple(x)
        x[random.choice([0,1])] += random.choice([-1,1])
```
